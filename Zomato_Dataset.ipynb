5. Cuisine Counts

```python
df['Cuisine'].value_counts()
```

6. Bar Graph

```python
df['Cuisine'].value_counts().plot(kind='bar')
plt.title("Restaurants Per Cuisine")
plt.xlabel("Cuisine")
plt.ylabel("Number of Restaurants")
plt.show()
```

7. Pie chart

```python
df['Online_Order'].value_counts().plot(kind='pie', autopct='%.2f%%')
plt.title("Online Order Availability")
plt.show()
```

8. Histogram

```python
plt.hist(df['Rating'], bins=5)
plt.title("Restaurant Rating Distribution")
plt.xlabel("Ratings")
plt.ylabel("Restaurants")
plt.show()
```

```python
sns.distplot(df['Rating'])
plt.show() to
```

10. Boxplot

```python
sns.boxplot(df['Rating'])
plt.title("Restaurant Ratings Boxplot")
plt.show()
```

11. Scatterplot

```python
sns.scatterplot(x=df['Delivery_Time'], y=df['Rating'])
plt.title("Delivery Time vs Rating")
plt.show()
```

12. Barplot

```python
sns.barplot(x=df['Cuisine'], y=df['Rating'], hue=df['Online_Order'])
plt.xticks(rotation=45)
plt.show()
```

13. Boxplot With Hue

```python
sns.boxplot(x=df['Online_Order'], y=df['Rating'], hue=df['Cuisine'])
plt.xticks(rotation=45)
plt.show()
```
